$min$$max$$scalar$

```python
# read the data again
path =r'C:\Users\aramaiah.ASUAD\Naresh_IT\MyDataScience\Data_Files\Visadataset.csv'
visa_df=pd.read_csv(path)
```

```python
# x-x_min/(x_min-x_min)
# step-1:Read the pwage column 
# step-2:Find the min value of the pwage column 
# step-3:Find the max value of the pwage column 
# step-4:nr=datacolumn-min value 
# step-5:dr=max_value-min_value
# step-6:nr/dr

# x=data column 
# x_min
# x_max
# nr=x-x_min
# dr=x_max-x_min
# nr/dr
```

```python
pwage=visa_df['prevailing_wage']
pwage_min=visa_df['prevailing_wage'].min()
pwage_max=visa_df['prevailing_wage'].max()
nr1=pwage-pwage_min
dr1=pwage_max-pwage_min
visa_df['prevailing_wage_normalization']=nr1/dr1
visa_df['prevailing_wage_normalization']
```
